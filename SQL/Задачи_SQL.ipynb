b)	В магазине проходит акция «Каждый  третий товар отдается бесплатно». Необходимо выстроить товары в чеке в порядке убывания цены и обнулить цену каждой третьей позиции. 

```
# запрос через оконную функцию

# считаем, что в чеке может быть разное число SKU и обнуляем каждый 3й товар по убыванию цены

sql = '''
with base_table as 
  (
  select 
    tab.*, row_number() over (partition by tab.tip order by tab.price desc) as number,
    case when (row_number() over (partition by tab.tip order by tab.price desc))%3=0 then 1 else 0 end as flag
  from tab
  )

select bt.*,
case when bt.flag=1 then 0 else bt.price end as new_price
from base_table bt
'''
```

```
select(sql)
```
